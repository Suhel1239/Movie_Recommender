# Item/Item Collaborative Filtering

```python
from sklearn.metrics.pairwise import cosine_similarity
# similar_score = cosine_similarity(movies_ratings_standarised.T)
# // Instead of fins=ding and tj=hen changing the matrix we can directly use pearson method
```

```python
# similar_score_df = pd.DataFrame(similar_score , index=movies_ratings_mat2.columns , columns=movies_ratings_mat2.columns )
similar_score_df = movies_ratings_mat2.corr(method="pearson")
```

```python
similar_score_df.head(5)
```

```python
type(similar_score_df)
```

```python
similar_score_df.info()
```

```python
for row in similar_score_df.index:
  print(row)
```

```python
def Similarity(moviename):
  similar_movies_score = similar_score_df[moviename]
  similar_movies_score = similar_movies_score.sort_values(ascending=False)
  return similar_movies_score.nlargest(10)
```

```python
x= Similarity("'burbs, The (1989)")
type(x)
```

```python
movies = ["'burbs, The (1989)"]
for movie in movies:
  x= Similarity(movie)
print(x)
```
